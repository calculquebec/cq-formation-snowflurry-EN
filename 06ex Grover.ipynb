# Grover algorithm

The Grover algorithm is to find a specific data point within an unstructured database. For example, find a specific card in a mixed deck of cards. 
L'algorithme de Grover résout le problème d'une recherche non structurée. Par exemple, trouver une carte précise dans un jeu de cartes mélangé. This is a quantum algorithm for finding the input value $x_0$ of an Oracle function $f(x)$ with $f(x_0)=1$ and $f(x)=0$ for all other values of $x$. The strength of the algorithm is that when the Oracle function has $N$ possible input values, the algorithm only requires $\mathcal{O}(\sqrt{N})$ evaluations from the Oracle, whereas the best classical algorithm requires $\mathcal{O}(N)$ evaluations.

## How the algorithm works
In this example, the initial states will be integers between $0$ and $2^n$. The $2^n$ states will be codified into the $|0\rangle$ and $|1\rangle$ states of the $n$ qubits. Next, we define an Oracle function that returns zero for all initial states, except for the desired value. The Oracle function is an operator (a series of quantum logic gates) $O$ such that $O|x\rangle = (-1)^(f(x))|x\rangle$. In other words, the Oracle sets the probability amplitude of the initial state $|x\rangle$ to zero if and only if $f(x) = 1$.

The example below uses 3 qubits, so the function $f(x)$ can take 8 values corresponding to the numbers 0 to 7 in binary. The objective is to find the number 6 which is 110 in binary.

The steps to follow:
1. Initialize the qubits to the $|0\rangle$ state and create a superposition of all states.
2. Run the Oracle function which cancels the probability magnitude of the input value if and only if the input is 110.
3. Apply Grover's diffusion operator. This operator averages the probability amplitudes and inverts the amplitudes around the mean so that the amplitude of the desired state is increased.
4. The circuit is repeated $\mathcal{O}(\sqrt{N})$ times, therefore 2 times
